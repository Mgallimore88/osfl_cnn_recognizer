The hawkears output is in files corresponding to audio clips. The input files are numbers, assumption is that the files are loaded in alphabetical order. try to load the files and match the predictions with the clips. 

```python
# imports
import glob
from pathlib import Path
import pandas as pd

# for utils
from pathlib import Path
import sys
BASE_PATH = Path.cwd().parent.parent
sys.path.append(str(BASE_PATH))
from src.utils import display_all
```

```python
data_path = Path.cwd().parent.parent / 'data'
hawkears_output_path_present = Path(data_path / 'processed' / 'hawkears_output'/ 'present')
hawkears_output_path_absent = Path(data_path / 'processed' / 'hawkears_output'/ 'absent')
```

```python
hawkears_present_output = glob.glob(str(hawkears_output_path_present / '*.txt'))
hawkears_absent_output = glob.glob(str(hawkears_output_path_absent / '*.txt'))
```

```python
present_results = []

for file_path in hawkears_present_output:

    with open(file_path, 'r') as f:
        file_content = f.read().strip().split('\n')
        file_id = file_path.split('/')[-1].split('.')[0]
        file_id = file_id.split('_')[0]
        for detection in file_content:
            if detection:
                line = [file_id] + detection.split('\t')
            else:
                line=[file_id]
            present_results.append(line)
        
            
df_present = pd.DataFrame(present_results, columns=['file_id', 'start_time', 'end_time', 'species;confidence'])
# df.file_id = df.file_id.astype(int)
# df.sort_values(by=['file_id'], inplace=True)
# df[['species', 'confidence']] = df['species;confidence'].str.split(';', expand=True)
display_all(df_present)
```

```python
absent_results = []

for file_path in hawkears_absent_output:

    with open(file_path, 'r') as f:
        file_content = f.read().strip().split('\n')
        file_id = file_path.split('/')[-1].split('.')[0]
        file_id = file_id.split('_')[0]
        for detection in file_content:
            if detection:
                line = [file_id] + detection.split('\t')
            else:
                line=[file_id]
            absent_results.append(line)
        
            
df_absent = pd.DataFrame(absent_results, columns=['file_id', 'start_time', 'end_time', 'species;confidence'])
df_absent[['species', 'confidence']] = df_absent['species;confidence'].str.split(';', expand=True)
display_all(df_absent)
```
